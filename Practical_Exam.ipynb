```python
#Given a list A of N distinct integer numbers, you can sort the list by moving an element to the end of the list. Find the minimum number 
#of moves required to sort the list using this method in ascending order

ls = [1,3,2,4,6,5]
c=0
i=0
while i < len(ls)-1:
  min=ls[i]
  val=ls[i]
  for x in range(i+1, len(ls)):
    val=ls[x]
    if val < min:
      break
  if val < min:
    ls.remove(min)
    ls.append(min)
    c+=1
  else:
    i+=1

  print(f'{c} moves are required')
```

```python
#Create class Triangle, Create object from it . The objects will have 3 attributes named a,b,c that represent sides of triangle .Triangle class 
# will have two methods init method to initialize the sides & method to calculate perimeter of triangle from its sides . Perimeter of triangle 
# should be printed from outside the class

class Triangle:
  def __init__(self,a,b,c):
    self.a=a;
    self.b=b;
    self.c=c;

  def perimeter(self):
    return self.a+self.b+self.c

a=int(input("a:"))
b=int(input("b:"))
c=int(input("c:"))

t=Triangle(a,b,c)

print(t.perimeter())
```

```python
#Write a program to create a class which performs basic calculator operations
class MathsOperations:
  def __init__(self,a,b):
    self.a=a
    self.b=b

  def add(self):
    res=self.a+self.b
    return res

  def sub(self):
    res=self.a-self.b
    return res

  def mul(self):
    res=self.a*self.b
    return res

  def div(self):
    res=self.a/self.b
    return res

a=int(input("Enter first number: "))
b=int(input("Enter second number: "))

obj=MathsOperations(a,b)

choice=1
while choice!=0:
  choice=int(input("Enter option: 1. Add 2. Sub 3. Mul 4. Div"))
  if choice==1:
    res=obj.add()
    print(res)
  elif choice==2:
    res=obj.sub()
    print(res)
  elif choice==3:
    res=obj.mul()
    print(res)
  elif choice==4:
    res=obj.div()
    print(res)
```

```python
#Write a Python class to reverse a string word by word
class revclass:
  def __init__(self,s):
    self.s=s
  def rev(self):
    self.s=' '.join(reversed(s.split())) #The join() method takes all items in an iterable and joins them into one string. ' ' is a seperator
                                            #split() spits a string into words
    return self.s

s="Hi I am Pusti"
r=revclass(s)
print(r.rev())
```

```python
#Write a Python class named Circle constructed by a radius and two methods which will compute the area and the perimeter of a circle.
class Circle:
  def __init__(self,r):
    self.r=r
  def area(self):
    return 3.14*self.r*self.r
  def perimeter(self):
    return 2*3.14*self.r

r=float(input("Enter radius of circle:"))
c=Circle(r)
a=c.area()
print(a)
p=c.perimeter()
print(p)
```

```python
# With given list L of integers , write a program that prints this L after removing duplicate values with original order preserved. 
#  Input: 12 24 35 24 88 120 155 88 120 155 
#  Output: 12 24 35 88 120 155

list1=input("Enter items:")
list1=list1.split()

duplicate=[]
unique=[]

for x in list1:
  if x not in duplicate:
    duplicate.append(x)
    unique.append(x)

print(" ".join(unique))
```

```python
#Given ab integer number n , define function named printDict(),which can print a dictionary where keys are numbers between 1 to n (both 
#included) and values are square of keys . Function printDict(), does not take any arguments

def printDict():
  dict={}
  n=int(input("Enter n:"))
  for i in range(1,n+1):
    dict[i]=i*i

  print(dict)

printDict()
```

```python
#Python program to generate the prime numbers from 1 to N

n=int(input("Enter the number till you want to check: "))

for i in range(1,n+1):
  if i>1:
    for j in range(2,i):
      if (i%j)==0:
        break
    else:
      print(i)
```

```python
#Python program to print the numbers from a given number n till 0 using recursion
n=int(input("Enter integer: "))

def printnum(n):
  if n==0:
    print(0)
  else:
    print(n)
    printnum(n-1)

printnum(n)
```

```python
#python class to convert a roman numeral to an integer

n=input("Enter roman number:")

dict={'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
res=0

for i in range(0,len(n)):
  if i>0 and dict[n[i]]>dict[n[i-1]]:
    res=res+dict[n[i]]-2*dict[n[i-1]]
  else:
    res=res+dict[n[i]]

print(res)
```

```python
#Write a Python class to get all possible unique subsets from a set of distinct integers
class Sol:
    def sub(self, sset):
        return self.subrecur([], sorted(sset))
    
    def subrecur(self, current, sset):
        if sset:
            return self.subrecur(current, sset[1:]) + self.subrecur(current + [sset[0]], sset[1:])
        return [current]

print(Sol().sub([6,5,4]))
```

```python
#Write a Python class to find a pair of elements (indices of the two numbers) from a given array whose sum equals a specific target 
#number
class Sol:
  def pairofele(self,list,target):
    dict={}
    for i,num in enumerate(list):
      if target-num in dict:
        return[dict[target-num],i]

      dict[num]=i

obj=Sol()
obj.pairofele([10,20,10,40,50,60,70],50)
```

```python
#Write a Python class to implement pow(x, n)
class Sol:
  def pow(self,x,n):
    if x==1 or x==0 or n==1:
      return x
    if x==-1:
      if n%2==0:
        return 1
      else:
        return -1

    if n==0:
        return 1

    if n<0:
      return 1/self.pow(x,-n)

    val=self.pow(x,n//2)

    if n%2==0:
      return val*val

    return val*val*x

obj=Sol()
obj.pow(2,2)
      
```
